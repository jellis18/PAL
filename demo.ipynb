## Running Noise Estimation ##

```python
# at the moment I have a MultiNest and an MCMC implementation of the Hybrid Time-Frequency Method (HTFM) although the MultiNest has more options at the moment
!python testing/lentati_single_pulsar_noise.py -h
```

```python
# run one of the MDC pulsars as an example
!python testing/lentati_single_pulsar_noise.py --h5File ./mdc_data/open1_demo.hdf5 --outDir demo_noise_results/powerlaw/J0030+0451/ --pulsar J0030+0451 --nmodes 10 --powerlaw
```

```python
# now take a look at the results
d = np.loadtxt('demo_noise_results/powerlaw/J0030+0451/testpost_equal_weights.dat')

# since this is the open MDC we know the "true" values [efac, log equad, log Agw, gw power spectral index]
inj = [1.0, -9, np.log10(5e-14), 4.333]
label = ['EFAC', 'log EQUAD', 'log Agw', 'gamma']

# call the triplot function from bayesutils to take a look at the 1 and 2 dimensional marginalized posteriors
bu.triplot(d[:,:-1], inj=inj, labels=label)
```

```python
# we can also do this with the model-independent method
!python testing/lentati_single_pulsar_noise.py --h5File ./mdc_data/open1_demo.hdf5 --outDir demo_noise_results/independent/J0030+0451/ --pulsar J0030+0451 --nmodes 10 
```

```python
# again we can make a triangle plot (it works for arbritrary size)
# now take a look at the results
d = np.loadtxt('demo_noise_results/independent/J0030+0451/testpost_equal_weights.dat')

# since this is the open MDC we know the "true" values 
label = ['EFAC', 'log EQUAD']
for ii in range(d.shape[1]-3):
    label.append(r'$\rho_%s$'%(int(ii)))
    
bu.triplot(d[:,0:-1], labels=label)
```

```python
# we can also plot the power spectrum
nmodes = 10
T = 5 * 3.16e7    # 5 year data span
f = np.linspace(1/T, nmodes/T, nmodes)    # nyquist sampling

# log of power spectrum coefficients from output
lp = np.mean(d[:,2:-1], axis=0)
err = np.std(d[:,2:-1], axis=0)

# make plot
plt.figure(figsize=(7,5))
plt.errorbar(np.log10(f), lp, err, fmt='o')
plt.grid()
plt.xlim(-8.0, -7.1)
plt.xlabel('log Frequency [Hz]')
plt.ylabel('log Power')
```

## Simulating Data ##

```python
# run help for PALSimulation
!python PALSimulation.py -h
```

```python
# first lets simulate a single source with SNR=15 and just white noise for the pulsars in open1_demo.hdf5
!python PALSimulation.py --h5File './mdc_data/open1_demo.hdf5' --single --snr 15 --gwchirpmass 2e8 --outFile ./mdc_data/open1_demo_single.hdf5
```

```python
# lets read the data into the pulsar class and take a look at the residuals
pfile = h5.File('./mdc_data/open1_demo_single.hdf5')

pulsargroup = pfile['Data']['Pulsars']

psr = [PALpulsarInit.pulsar(pulsargroup[key], addGmatrix=True, addNoise=True) for key in pulsargroup]
```

```python
# plot it
figure(figsize=(11.5, 8))
for ct, p in enumerate(psr):
    plt.subplot(6, 6, ct+1)
    plt.errorbar(p.toas/86400, p.res*1e6, p.err*1e6, fmt='.')
    plt.grid()
    
plt.tight_layout(h_pad=0.1, w_pad=0.1)
```

```python
# now lets simulate a stochastic background with Agw = 5e-15 with white noise
!python PALSimulation.py --h5File './mdc_data/open1_demo.hdf5' --outFile './mdc_data/open1_demo_gwb.hdf5' --gwb  --gwbAmp 5e-15
```

```python
# again lets read in the data 
# lets read the data into the pulsar class and take a look at the residuals
pfile = h5.File('./mdc_data/open1_demo_gwb.hdf5')

pulsargroup = pfile['Data']['Pulsars']

psr = [PALpulsarInit.pulsar(pulsargroup[key], addGmatrix=True, addNoise=True) for key in pulsargroup]
```

```python
# and plot it

figure(figsize=(11.5, 8))
for ct, p in enumerate(psr):
    plt.subplot(6, 6, ct+1)
    plt.errorbar(p.toas/86400, p.res*1e6, p.err*1e6, fmt='.')
    plt.grid()
    
plt.tight_layout(h_pad=0.1, w_pad=0.1)
```
